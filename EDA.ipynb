```python
#2check duplicates in UPCImages.csv
import pandas as pd

# Assuming your dataset is in a CSV file named 'dataset.csv'
df = pd.read_csv('UPCImages.csv')

# Count occurrences of each ID
id_counts = df['ItemSku'].value_counts()

# Filter IDs that appear more than 8 times
ids_appearing_more_than_8_times = id_counts[id_counts > 8]
ids_appearing_more_than_16_times = id_counts[id_counts > 16]

print("IDs appearing more than 8 times and their counts:")
print(ids_appearing_more_than_8_times)
# print(ids_appearing_more_than_16_times)

total_counts_8 = len(ids_appearing_more_than_8_times)
total_counts_16 = len(ids_appearing_more_than_16_times)

print("Total counts of IDs appearing more than 8 times:", total_counts_8)
print("Total counts of IDs appearing more than 16 times:", total_counts_16)


#remove duplicates
df_image = df.drop_duplicates(subset=['ItemSku'])
# Count null values in each column
null_counts = df_image.isnull().sum()

# Total number of rows in the dataframe
total_rows = len(df_image)
counts = total_rows - df_image.isnull().sum()

# Calculate percentage of null values in each column
null_percentage = (null_counts / total_rows) * 100

# Create a new dataframe to display the results
null_df = pd.DataFrame({'Column': null_counts.index, 'Null Count': null_counts.values, 'Null Percentage': null_percentage.values, 'Non-Null Count': counts.values})

print(null_df)
print("Total counts of unique IDs:", total_rows)
```
